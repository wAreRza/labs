```python
sns.histplot(df['AGE_50'], kde=True)
plt.title('Распределение AGE_50')
plt.xlabel('AGE_50')
plt.ylabel('Частота')
plt.show()
```

 Большинство пациентов находятся в диапазоне 43–53 лет

```python
sns.countplot(x='MD_50', data=df)
plt.title('Распределение MD_50')
plt.xlabel('MD_50')
plt.ylabel('Количество')
plt.show()
```

Категория 3 является наиболее распространённой. Можно предположить, что она отражает умеренное состояние, которое встречается чаще всего.

```python
sns.boxplot(x=df['SBP_50'])
plt.title('Boxplot SBP_50')
plt.xlabel('SBP_50')
plt.show()
```

Медиана равна 120. Выбросы показывают наличие пациентов с аномально высоким или низким давлением
